```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
col=['sepal.length','sepal.width','petal.length','petal.width','variety']
iris=pd.read_csv("iris.csv",names=col)

print("shape:",iris.shape)
print("\nFirst five rows\n")
print(iris.head())

print("\nLast five rows\n")
print(iris.tail())

print("Size of dataset:",iris.size)

print("\nno of samples available for each type\n") 
print(iris["variety"].value_counts())

print("\n Description \n",iris.describe())
```

```python
iris = sns.load_dataset('iris')

sns.pairplot(iris, kind="kde",hue="species", markers='.')
sns.pairplot(iris, kind="scatter",hue="species")
sns.pairplot(iris, kind="hist",hue="species",markers='o')
sns.pairplot(iris, kind="reg",hue="species",markers='+')
```

```python

df = sns.load_dataset('iris')
sns.displot(x = 'sepal_length',hue="species" , data =df)
```

```python
df = sns.load_dataset('iris')
sns.histplot(x = 'sepal_width',hue="species" , data =df)
```

```python
df = sns.load_dataset('iris')
sns.relplot(x = 'sepal_width',y='sepal_length',hue="species", data =df)
```
